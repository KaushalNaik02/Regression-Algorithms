### Implementing Multiple Linear Regression Using Gradient Descent

### In this MLR we solve The bias abd weights differnrtly(theta0 and Theta1,2,3,..  has differnt Formula )

```python
import numpy as np

# Sample data (replace with your actual data)
# X should be a 2D array where each row is a sample and each column is a feature
# y should be a 1D array of target values
X = np.array([[1, 1], [2, 2], [3, 3], [4, 4]])
y = np.array([10, 20, 30, 40])

```

```python
# Hyperparameters
learning_rate = 0.01
n_iterations = 2201

```

```python
# Initialize parameters
n_samples, n_features = X.shape
weights = np.zeros(n_features)
bias = 0
print(n_samples)
print(n_features)
print(weights)
print(bias)
```

```python
# Gradient descent
for iteration in range(n_iterations):
    # Calculate predictions
    y_predicted = np.dot(X, weights) + bias
    print(y_predicted)
    
    # Compute gradients
    dw = (1/n_samples) * np.dot(X.T, (y_predicted - y))
    db = (1/n_samples) * np.sum(y_predicted - y)
    
    # Update parameters
    weights -= learning_rate * dw
    bias -= learning_rate * db
    print(weights)
    print(bias)
    # Calculate and print cost every 100 iterations (optional)
    if iteration % 100 == 0:
        cost = (1/(2*n_samples)) * np.sum((y_predicted - y)**2)
        print(f"Iteration {iteration}, Cost: {cost:.4f}")

```

```python
# Print final parameters
print("\nFinal parameters:")
print("Weights:", weights)
print("Bias:", bias)

# Make predictions
X_test = np.array([[8, 8]])
predictions = np.dot(X_test, weights) + bias
print("\nPredictions for test data:")
print(predictions)
```
